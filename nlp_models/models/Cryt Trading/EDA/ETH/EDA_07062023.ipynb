```python
import pandas as pd
import numpy as np
import datetime
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df_ETH = pd.read_csv('final_ETHUSDT.csv')
```

```python
df_ETH.head()
```

```python
df_ETH['open_time']
```

```python
df_ETH['datetime'] = df_ETH['open_time'].apply(lambda x: datetime.datetime.strptime(x, "%m/%d/%Y, %H:%M:%S"))
```

```python
df_ETH['year'] = df_ETH['datetime'].apply(lambda x: x.year)
df_ETH['month'] = df_ETH['datetime'].apply(lambda x: x.month)
df_ETH['day'] = df_ETH['datetime'].apply(lambda x: x.day)
df_ETH['day_of_week'] = df_ETH['datetime'].apply(lambda x: x.weekday())
```

```python
df_ETH.head()
```

```python
df_ETH.dtypes
```

```python
df_ETH = df_ETH.sort_values('datetime', ascending = True)
```

```python
df_ETH['change_open'] = - (df_ETH['open'].shift(1) - df_ETH['open'])
```

```python
df_ETH.head()
```

```python
df_ETH = df_ETH.drop('ignore', axis =1)
```

```python
corr_matrix = df_ETH.corr()
```

```python
# sn.heatmap(corr_matrix, annot=True)
# plt.show()
```

```python
mask = np.zeros_like(corr_matrix)
mask[np.triu_indices_from(mask)] = True
with sn.axes_style("white"):
    f, ax = plt.subplots(figsize=(15, 11))
    ax = sn.heatmap(corr_matrix, mask=mask, vmax=1.0, vmin= -1.0, square=True, annot=True)
    bottom, top = ax.get_ylim()
    ax.set_ylim(bottom + 0.5, top - 0.5)
    
plt.show()
```

### Load Twitter data

```python
df_tweet = pd.read_csv('ETH_tweets_cleaned.csv')
df_tweet.head()
```

```python
df_tweet_senti = pd.get_dummies(df_tweet.sentiment, prefix='Sentiment')
# df_tweet_senti.shape
df_tweet = df_tweet.join(df_tweet_senti)
df_tweet.shape
```

```python
df_tweet_agg = df_tweet.groupby(['year','month','day']).agg({'Sentiment_NEGATIVE':'sum','Sentiment_NEUTRAL':'sum',
                                                            'Sentiment_POSITIVE':'sum'}).reset_index()

df_tweet_agg.head(2)
```

```python
df_tweet_agg.shape
```

```python
df_ETH.shape
```

```python
df_joined = pd.merge(df_ETH, df_tweet_agg, how='left', on=['year','month','day'])

df_joined.shape
```

```python
df_joined.head(2)
```

```python
#filter for the days twitter data is avialable
df_filter = df_joined[df_joined['datetime'] >= datetime.datetime(2023, 6, 1)] #TODO rewrite
```

```python
df_filter.dtypes
```

```python
#Aggregating the dataframe by day to find out any correlation between the tweets and the prices 

df_daily = df_filter.groupby(['day','month','year','day_of_week']).agg({'open_time':'min',
                                                                         'close_time':'max',
                                                                         'open':'first',
                                                                         'high':'max',
                                                                         'low':'min',
                                                                         'close':'last',
                                                                         'volume':'sum',
                                                                         'num_trades':'sum',
                                                                         'open_timestamp':'min',
                                                                         'close_timestamp':'max',
                                                                         'Sentiment_NEGATIVE':'mean',
                                                                         'Sentiment_NEUTRAL':'mean',
                                                                         'Sentiment_POSITIVE':'mean'}).reset_index()


df_daily['VolperTrade'] = df_daily['volume']/df_daily['num_trades']
df_daily.head()
```

```python
df_daily.columns
```

```python
corr = df_daily[['day_of_week', 'open_time', 'close_time',
               'open', 'high', 'low', 'close', 'volume', 'num_trades',
               'open_timestamp', 'close_timestamp', 'Sentiment_NEGATIVE',
               'Sentiment_NEUTRAL', 'Sentiment_POSITIVE', 'VolperTrade']].corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sn.axes_style("white"):
    f, ax = plt.subplots(figsize=(12, 9))
    ax = sn.heatmap(corr, mask=mask, vmax=1.0, vmin= -1.0, square=True, annot=True)
    bottom, top = ax.get_ylim()
    ax.set_ylim(bottom + 0.5, top - 0.5)
    
plt.show()
```
